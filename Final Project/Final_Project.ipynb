```python
# 05-09/04/2021
# TAYFUN
# ÇINARLI
# tayfuncinarli@gmail.com
# PROJECT

class Soru:
     def __init__(self, giris, cevap):
          self.giris = giris
          self.cevap = cevap
sorular = [
    "1- What color are kiwis?\na) Brown\nb) Orange\n",
    "2- What color are avocados?\na) Green\nb) Orange\n",
    "3- What color are Oranges?\na) Blue\nb) Orange\n",
    "4- What color are apples?\na) Red/Green\nb) Orange\n",
    "5- What color are strawberries?\na) Red\nb) Green\n",
    "6- What color are melons?\na) Orange\nb) Yellow\n",
    "7- What color are carrots?\na) Green\nb) Orange\n",
    "8- What color are plums?\na) Red\nb) Green\n",
    "9- What color are cherries?\na) Burgundy\nb) Orange\n",
    "10- What color are bananas?\na) Red/Green\nb) Yellow\n",
]
cevaplar = [
    Soru(sorular[0], "a"),
    Soru(sorular[1], "a"),
    Soru(sorular[2], "b"),
    Soru(sorular[3], "a"),
    Soru(sorular[4], "a"),
    Soru(sorular[5], "b"),
    Soru(sorular[6], "b"),
    Soru(sorular[7], "b"),
    Soru(sorular[8], "a"),
    Soru(sorular[9], "b")]

skor = 0
for soru in cevaplar:
    cevap = input(soru.giris)
    if cevap.lower() == soru.cevap.lower():
        skor += 1
if skor>5:
    print("BAŞARILI!:) ",skor*10,"PUAN\n",len(cevaplar)," soruda ", skor, "doğru cevap verdiniz.")
else:
    print("BAŞARISIZ! :( ",skor*10,"PUAN\n",len(cevaplar)," soruda ", skor, " doğru cevap verdiniz.")
```
